```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Train (1).csv')
df.head()
```

# VISUALIZATION

```python
sns.set_style('whitegrid')
```

CATEGORICAL FEATURES

```python
plt.figure(figsize=(18,18))
plt.subplot(3,3,1)
sns.countplot(x='Warehouse_block',data=df)
plt.title('Warehouse Block',fontsize=15)

plt.subplot(3,3,2)
sns.countplot(x='Mode_of_Shipment',data=df)
plt.title('Mode of Shipment',fontsize=15)

plt.subplot(3,3,3)
sns.countplot(x='Customer_care_calls',data=df)
plt.title('Customer Care Calls',fontsize=15)

plt.subplot(3,3,4)
sns.countplot(x='Customer_rating',data=df)
plt.title('Customer Rating',fontsize=15)

plt.subplot(3,3,5)
sns.countplot(x='Prior_purchases',data=df)
plt.title('Prior Purchases',fontsize=15)

plt.subplot(3,3,6)
sns.countplot(x='Product_importance',data=df)
plt.title('Product Importance',fontsize=15)

plt.subplot(3,3,7)
sns.countplot(x='Gender',data=df)
plt.title('Gender',fontsize=15)
```

1. We see that Warehouse F has more product than the other.
2. Most of the product are transported through ship.
3. There are less product with high product importance.
4. The product purchased is slightly more in  Female.

HISTOGRAM/DISTRIBUTION OF NUMERIC FEATURES

```python
plt.figure(figsize=(8,5))
plt.hist(df.Cost_of_the_Product,bins=20)
plt.title('Cost of the Product',fontsize=15)
```

1) There are high no. of product which costs between 260-270 $

```python
plt.figure(figsize=(8,5))
plt.hist(df.Discount_offered,bins=20)
plt.title('Discount Offered',fontsize=15)
```

There are maximum product there the discount offered between 0-10 %.

```python
plt.figure(figsize=(8,5))
plt.hist(df.Weight_in_gms,bins=20)
plt.title('Weight in Gms',fontsize=15)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(3,3,1)
sns.countplot(x='Reached.on.Time_Y.N',data=df)
plt.title('Reached ON Time',fontsize=15)
```

More number product reached on time compared to which didn't.

```python
BOXPLOT OF NUMERIC FEATURES/ TO CHECK THE OUTLIERS
```

```python
plt.figure(figsize=(12,12))

plt.subplot(2,3,1)
plt.boxplot(df.Cost_of_the_Product)
plt.title('Cost of the Product',fontsize=15)

plt.subplot(2,3,2)
plt.boxplot(df.Discount_offered)
plt.title('Discount offered',fontsize=15)

plt.subplot(2,3,3)
plt.boxplot(df.Weight_in_gms)
plt.title('Weight in gms',fontsize=15)

plt.subplot(2,3,4)
plt.boxplot(df.Customer_care_calls)
plt.title('Customer Care Calls',fontsize=15)

plt.subplot(2,3,5)
plt.boxplot(df.Customer_rating)
plt.title('Customer rating',fontsize=15)

plt.subplot(2,3,6)
plt.boxplot(df.Prior_purchases)
plt.title('Prior Purchases',fontsize=15)
```

From the figure we see there are outlier in the DISCOUNT OFFERED and PRIOR PURCHASES. I have removed the outlier in my Data Cleaning section.

CORRELATION OF THE VARIABLES

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,fmt='.2f',cmap='Reds')
```

Based on the Correlation heatmap:
1. Reached on time has positive correlation with Discount Offered and weak negative correlation with Weight in Gms.
2. Customer Care calls has postive correlation with Cost of the Product.
3. Discount Offered has negative correlation with Weight in Gms
