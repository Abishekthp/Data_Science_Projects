```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('telecom_users.csv')
df.head()
```

```python
cols=['OnlineBackup','StreamingTV','StreamingMovies','DeviceProtection','TechSupport','OnlineSecurity']
for i in cols:
    df[i]=df[i].replace({'No internet service':'No'})
```

```python
df.info()
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.Churn=df.Churn.apply(lambda x: 1 if x=='Yes' else 0)
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.drop('customerID',axis=1,inplace=True)
df
```

```python
new_df=pd.get_dummies(df,drop_first=True)
new_df
```

```python
new_df.describe()
```

```python
new_df.tenure=new_df.tenure/100
new_df.MonthlyCharges=new_df.MonthlyCharges/150
new_df.TotalCharges=new_df.TotalCharges/9500
```

```python
new_df.describe()
```

```python
y = new_df['Churn']
X = new_df.drop('Churn',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)*100 , model.score(X_train,y_train)*100
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,model.predict(X_test))
cr
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model.predict(X_test))
```

# DECISION TREE

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
tree.score(X_test,y_test)*100 , tree.score(X_train,y_train)*100
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,tree.predict(X_test))
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=25)
```

```python
forest.fit(X_test,y_test)
```

```python
forest.score(X_test,y_test)*100 , forest.score(X_train,y_train)*100
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,forest.predict(X_test))
```
