```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

# DATA CLEANING OF TRAIN DATA

```python
df_train.head()
```

REMOVE PUNCTUATION

```python
import string
string.punctuation
```

```python
def remove_punc(text):
    text_no_punc="".join([char for char in text if char not in string.punctuation])
    return text_no_punc
df_train['removed_punc']=df_train['tweet'].apply(lambda x: remove_punc(x))
```

```python
df_train.head()
```

TOKENIZE TEXT

```python
import re
```

```python
def tokenize(text):
    tokens=re.split('\W+',text)
    return tokens
df_train['tokenized_tweet']=df_train['removed_punc'].apply(lambda x: tokenize(x.lower()))
```

```python
df_train.head()
```

REMOVE STOPWORDS

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text_no_stopwords = [word for word in text if word not in stopwords]
    return text_no_stopwords
df_train['no_stopwords_tweet']=df_train['tokenized_tweet'].apply(lambda x: remove_stopwords(x))
```

```python
df_train.head()
```

LEMMATIZING

```python
import nltk
wn=nltk.WordNetLemmatizer()
```

```python
def lemmatize(text):
    text_lem = [wn.lemmatize(word) for word in text]
    return text_lem
df_train['lemmatized_text']=df_train['no_stopwords_tweet'].apply(lambda x: lemmatize(x))
```

```python
df_train.head()
```

```python
df_train.drop(['id','tweet','removed_punc','tokenized_tweet','no_stopwords_tweet'],axis=1,inplace=True)
```

```python
df_train
```

```python
df_train.rename({'lemmatized_text':'cleaned_tweet'},axis=1)
```
